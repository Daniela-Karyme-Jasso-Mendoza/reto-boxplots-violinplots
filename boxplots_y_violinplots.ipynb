````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import userdata
````

````python
ruta = userdata.get('ruta')

df = pd.read_csv(ruta + 'melbourne_housing-clean.csv', index_col=0)

df.head()
````

````python
plt.figure(figsize=(8, 10))
sns.boxplot(data=df, x='type', y='price', color='purple')
plt.title('Boxplot sobre tipos de Propiedad y Costos', fontsize=16)
plt.xlabel('Tipo de Propiedad')
plt.ylabel('Precio')
plt.show()
````

````python
plt.figure(figsize=(8, 10))
sns.violinplot(data=df, x='type', y='price', color='purple')
plt.title('Violinplot sobre tipos de Propiedad y Costos', fontsize=16)
plt.xlabel('Tipo de Propiedad')
plt.ylabel('Precio')
plt.show()
````
